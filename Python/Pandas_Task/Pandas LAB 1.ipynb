```python

df4 = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df4
```

```python
df4.set_index('customerID', inplace=True)
df4
```

```python
df4.info()
```

```python
df4['TotalCharges'] = pd.to_numeric(df4['TotalCharges'], errors='coerce')
df4
```

```python
df4.info()
```

```python
df4['TotalCharges'].fillna(df4['TotalCharges'].mean(), inplace=True)
df4
```

```python
df4.info()
```

```python
categorical_columns = df4.select_dtypes(include=['object']).columns
columns_to_encode = [col for col in categorical_columns if col not in ['Contract', 'Churn']]

df4_encoded = pd.get_dummies(df4, columns=columns_to_encode)
df4_encoded
```

```python
print(df4['Contract'].value_counts())

contract_mapping = {'Month-to-month': 1, 'One year': 2, 'Two year': 3}
#df4['Contract'] = df4['Contract'].map(contract_mapping)
df4['Contract'] = df4['Contract'].replace(contract_mapping)
df4
```

```python
df4.info()
```

```python
df4.drop_duplicates(inplace=True)

df4['Churn'] = df4['Churn'].replace({'No': 0, 'Yes': 1})

df4
```

```python
df4.info()
```

```python
df4['Contract'].fillna(df4['Contract'].mode()[0], inplace=True)
df4['Churn'].fillna(df4['Churn'].mode()[0], inplace=True)
df4
```

```python
df4.info()
```

```python
df4_churn = df4[['Churn']].copy()

df4.drop(columns=['Churn'], inplace=True)

numerical_columns = df4.select_dtypes(include=['float64', 'int64']).columns
df4[numerical_columns] = (df4[numerical_columns] - df4[numerical_columns].min()) / (df4[numerical_columns].max() - df4[numerical_columns].min())

df4.to_csv('df4_normalized.csv')
df4_churn.to_csv('df4_churn.csv')
```
