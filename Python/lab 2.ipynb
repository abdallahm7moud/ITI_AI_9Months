Write a program that counts up the number of vowels [a, e, i , o,
u] contained in the string.

```python
strr= input("Enter a string: ")
vowels = "aeiou"
count = 0
for char in strr.lower():
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```

Write a program that prints the number of times the string ' iti '
occurs in anystring

```python
strr = input("Enter a string: ")
iti = 'iti'
strr = strr.lower()
count = strr.count(iti)

print(f"The number of times '{iti}' occurs in the string is:", count)
```

Write a program that remove all vowels from the input word and
generate a brief version of it.

```python
strr = input("Enter a string: ")
vowels = "aeiou"
strr = strr.lower()
for char in strr:
    if char in vowels:
        strr = strr.replace(char, '')

print("Brief version of the word:", strr)
```

Write a program that prints the locations of " i " character in any
string you added.


```python
strr = input("Enter a string: ")
locations = []

for index, char in enumerate(strr):
    if char == 'i':
        locations.append(index)

print("Positions of 'i':", locations)
```

Write a program that generate a multiplication table from 1 to the
number passed.



```python
num =input("Enter a number: ") 
if num.isdigit():
    outer_list = []
    for i in range(1, int(num) + 1):
        inner_list = []
        for j in range(1, i + 1):
            inner_list.append(i * j)
        outer_list.append(inner_list)

    print(outer_list)
else:
    print("Invalid input")
```


Write a program that build a Mario pyramid like below:



```python
num = input("Enter the number: ")
if num.isdigit():
    num = int(num)
    for i in range(1, num + 1):
        print(" " * (num - i) + "*" * i)
else:
    print("Enter a valid number")
```
