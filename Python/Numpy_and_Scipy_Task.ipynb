```python
def L1(yhat, y):
    """
    Arguments:
    yhat -- vector of size m (predicted labels)
    y -- vector of size m (true labels)
    
    Returns:
    loss -- the value of the L1 loss function defined above
    """
    
    loss = loss = np.sum(np.abs(y - yhat))
    
    return loss



yhat = np.array([.9, 0.2, 0.1, .4, .9])
y    = np.array([1 ,  0 ,  0 ,  1, 1])

l1_loss = L1(yhat,y)
print("L1 = " + str(l1_loss))
```

```python
def L2(yhat, y):
    """
    Arguments:
    yhat -- vector of size m (predicted labels)
    y -- vector of size m (true labels)
    
    Returns:
    loss -- the value of the L2 loss function defined above
    """
    
    loss = np.sum(np.power(y - yhat, 2))
    
    return loss



yhat = np.array([.9, 0.2, 0.1, .4, .9])
y    = np.array([1 ,  0 ,  0 ,  1, 1])

l2_loss = L2(yhat,y)

print("L2 = " + str(l2_loss))
```

```python
import numpy as np
import time


def dot_product_for_loop(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result


T = 1000
a = np.random.randn(10_000)
b = np.random.randn(10_000)


start = time.process_time()
for t in range(T):
    dot_product_for_loop(a, b) 
end = time.process_time()
time_for_loop = end - start


print("Time For loop: ", time_for_loop)
```
