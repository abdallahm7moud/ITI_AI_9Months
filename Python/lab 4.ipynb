The aim of the project: Create a console app to start fundraise projects.
The app should include the following features:
1 - Authentication System:
• Registration:
• First name
• Last name
• Email
• Password
• Confirm password
• Mobile phone [validated against Egyptian phone numbers]
• Login
• The user should be able to login after activation using his email and password

Authentication System:


```python
import re
import os
import json
import time


user_data = {}

def validate_phone(phone_number):
    pattern = re.compile(r"^01[0-2,5]{1}[0-9]{8}$")
    return pattern.match(phone_number)

def validate_email(email):
    pattern = re.compile(r"^[a-zA-Z0-9]+@[a-zA-Z]+\.[a-zA-Z]+$")
    return pattern.match(email)

def validate_password(password):
    pattern = re.compile(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[!@#\$%\^&\*])(?=.{8,})")
    return pattern.match(password)

def ask_for_phone_number():
    phone_number = input("Enter phone number: ")
    if not validate_phone(phone_number):
        print("Invalid phone number. Please enter a valid Egyptian phone number.")
        ask_for_phone_number()
    else:
        return phone_number

def ask_for_email():
    email = input("Enter email: ")
    if email in user_data:
        print("Email already registered.")
        ask_for_email()
    else:
        if not validate_email(email):
            print("Invalid email. Please enter a valid email.")
            ask_for_email()
        else:
            return email
    
def ask_for_password():
    password = input("Enter password: ")
    if not validate_password(password):
        print("Invalid password. Please enter a valid password.")
        ask_for_password()
    else:
        confirm_password = input("Confirm password: ")
        if password != confirm_password:
            print("Passwords do not match. Please try again.")
            ask_for_password()
        else:
            return password
        
def save_data():
    with open("users.json", "w") as file:
        json.dump(user_data, file)

def load_data():
    global user_data
    try:
        with open("users.json", "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return {}
    
def login():
    global user_data
    user_data = load_data()
    os.system('cls')
    print("-----Welcome to the Login page------")
    Login_email = input("Enter email: ").strip()
    Login_password = input("Enter password: ").strip()
    user = user_data.get(Login_email)
    if not user:
        print("No account found with this email.")
        login()

    if user["password"] != Login_password:
        print("Incorrect password.")
        login()
    
    print("Login successful")
    print(f"Welcome back, {user['first_name']}!")
    
def register():
    global user_data
    os.system('cls')
    print("-----Welcome to the registration page-----")
    first_name = input("Enter first name: ").strip()
    last_name = input("Enter last name: ").strip()
    registration_email = ask_for_email().strip()        
    phone = ask_for_phone_number().strip()
    registration_password = ask_for_password()
    user_data[registration_email] = {
        "first_name": first_name,
        "last_name": last_name,
        "password": registration_password,
        "phone": phone
    }
    save_data()
    print("Registration successful")
    
    
def opthion():
    print("1-register")
    print("2-login")
    print("3-exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        register()
        time.sleep(5)
        os.system('cls')
        opthion() 
    elif choice == "2":
        login()
        time.sleep(5)
        os.system('cls')
        opthion()
    else:
        print("Goodbye!")
        return


if __name__ == "__main__":
    opthion()
```
